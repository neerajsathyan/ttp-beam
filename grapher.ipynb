```python
import pandas
```

```python
df = pandas.read_csv('data_sampled.csv')
```

```python
df.head(5)
```

```python
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')
ax.scatter3D(df['beam_width'], df['streak_limit'], df['shortest_path'])
ax.set_xlabel('Beam Width')
ax.set_ylabel('Streak Limit')
ax.set_zlabel('Shortest Path')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')
ax.scatter3D(df['beam_width'], df['streak_limit'], df['time_elapsed'])
ax.set_xlabel('Beam Width')
ax.set_ylabel('Streak Limit')
ax.set_zlabel('Time Elapsed')
plt.show()
```
